```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fetch price data
ticker = 'BTC-USD'
data = yf.download(ticker, interval='1h', period='30d')
prices = data['Close'].dropna()

# Calculate log(price) and log(return)
log_price = np.log(prices)
log_return = np.log(prices / prices.shift(1)).dropna()

# Compute statistics
mu_price = log_price.mean()[0]
sigma_price = log_price.std()[0]

mu_return = log_return.mean()[0]
sigma_return = log_return.std()[0]


# Plotting
plt.figure(figsize=(14, 6))

# Log(price) plot
plt.subplot(1, 2, 1)
sns.histplot(log_price, bins=50, kde=True, color='skyblue')
plt.title("Distribution of log(price)")
plt.xlabel("log(price)")

# Annotate  and 
plt.text(x=log_price.min(), y=plt.ylim()[1]*0.9,
         s=f"$\\mu$ = {mu_price:.4f}\n$\\sigma$ = {sigma_price:.4f}",
         fontsize=12, bbox=dict(facecolor='white', alpha=0.7))

# Log(return) plot
plt.subplot(1, 2, 2)
sns.histplot(log_return, bins=50, kde=True, color='salmon')
plt.title("Distribution of log(return)")
plt.xlabel("log(return)")

# Annotate  and 
plt.text(x=log_return.min(), y=plt.ylim()[1]*0.9,
         s=f"$\\mu$ = {mu_return:.4f}\n$\\sigma$ = {sigma_return:.4f}",
         fontsize=12, bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.savefig("log_price_and_return_distribution_1hr_30d.png", dpi=300)
plt.show()
```
