```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Define AR(3) process: Y_t = 0.6 Y_{t-1} - 0.3 Y_{t-2} + 0.2 Y_{t-3} + _t
ar_params = np.array([1, -0.6, 0.3, -0.2])  # AR coefficients (with sign change)
ma_params = np.array([1])  # No MA terms

# Create an AR(3) process
ar_process = ArmaProcess(ar_params, ma_params)
simulated_data = ar_process.generate_sample(nsample=300)  # Generate 300 data points

# Plot the AR(3) Time Series
plt.figure(figsize=(12, 4))
plt.plot(simulated_data, label="AR(3) Process")
plt.title("Simulated AR(3) Time Series")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.show()

# Plot ACF
plt.figure(figsize=(12, 4))
plot_acf(simulated_data, lags=30)
plt.show()

# Plot PACF
plt.figure(figsize=(12, 4))
plot_pacf(simulated_data, lags=30)
plt.show()
```
